```python
import numpy as np
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

```python
file = open("DDPG/models/ddpg_2.csv")
reader = csv.reader(file)
results = [row for row in reader]
```

```python

y1 = [float(result[1]) for result in results]
x1 = range(1, len(y1)+1)
x1 = np.multiply(x1, 100)
plt.plot(x1,y1)
```

```python
file2 = open("dqn/models/dqn.csv")
results = []
for line in file2.readlines():
    results.append(line.rstrip("\n"))
x = [int(res.split(",")[0]) for res in results][:5000:100]
y = [float(res.split(",")[1]) for res in results][:5000]
y_ave = []
for i in range(0, len(y), 100):
    y_ave.append(np.mean(y[i:i+99]))
plt.plot(x,y_ave)
```

```python
plt.plot(x1, y1, x, y_ave)
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.legend(["DDPG", "DQN"])
plt.grid()
plt.title("Model-Free Reinforcement Learning Results")
```
